```python
import os
import csv
import pandas as pd
```

```python
#Define the folder path where the raw data is located
folder_path_1 = r"C:\Users\tigra\OneDrive\Documents\GitHub\bearings-project-meng\Raw Data - 1st Test"
```

```python
#Make a path where the output CSV will be stored
output_file_1 = r"C:\Users\tigra\OneDrive\Documents\GitHub\bearings-project-meng\output_data_test1.csv"
```

```python
files = sorted(os.listdir(folder_path_1))

with open(output_file_1, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    
    # Loop through every second file in the folder
    for idx, filename in enumerate(files):
        if idx % 2 == 0:  # Skip every second file (write only files with even indices)
            file_path = os.path.join(folder_path_1, filename)

            file_number = int(idx / 2)
            print('Processing file # ' + str(file_number))
            
            # Open the text file and read its contents
            with open(file_path, 'r') as file:
                
                for line in file:
                    columns = line.strip().split() 
                    csvwriter.writerow(columns)  # Write the columns to the CSV

print(f"Data from every second text file has been successfully exported to {output_file_1}")
```
